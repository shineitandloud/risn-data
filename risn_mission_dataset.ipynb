```python

# Install required libraries
!pip install Pillow numpy torch torchvision scipy

# Import libraries
import os
import numpy as np
from PIL import Image, ImageDraw, ImageFilter
import random
import torch
import torchvision.transforms as transforms
from scipy.ndimage import gaussian_filter

# === CONFIGURATION ===
NUM_IMAGES = 10000
IMAGE_SIZE = (512, 512)
OUTPUT_DIR = "risn_mission_dataset"

# === ATTRIBUTE SETUP ===
shirt_types = ["short_sleeve", "button_down"]
patterns = ["stripes", "polka_dots", "florals", "solid", "camouflage"]
colors = ["red", "blue", "green", "yellow", "black", "white", "gradient", "metallic"]
materials = ["cotton", "polyester", "linen", "silk"]
ai_textures = ["holographic", "iridescent", "3d_printed"]

# === DATA LOADER FUNCTION ===
def load_pattern_texture(pattern_name):
    """Load texture for a given pattern"""
    try:
        texture_path = f"textures/{pattern_name}.png"
        return Image.open(texture_path).resize(IMAGE_SIZE)
    except:
        return None  # Default to solid if texture is missing

# === IMAGE GENERATOR FUNCTION ===
def generate_shirt_image(image_path, pattern, color, material, ai_texture):
    """Generate a synthetic shirt image with textures"""
    img = Image.new("RGB", IMAGE_SIZE, (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)))
    draw = ImageDraw.Draw(img)

    # Apply pattern
    pattern_texture = load_pattern_texture(pattern)
    if pattern_texture:
        img.paste(pattern_texture, (0, 0), pattern_texture)

    # Apply AI-driven texture effect
    if ai_texture == "holographic":
        img = img.filter(ImageFilter.EDGE_ENHANCE)
    elif ai_texture == "iridescent":
        img = img.filter(ImageFilter.EMBOSS)
    elif ai_texture == "3d_printed":
        img = img.filter(ImageFilter.DETAIL)

    # Save image
    img.save(image_path)

# === MAIN SCRIPT ===
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

for i in range(NUM_IMAGES):
    pattern = random.choice(patterns)
    color = random.choice(colors)
    material = random.choice(materials)
    ai_texture = random.choice(ai_textures)

    image_path = os.path.join(OUTPUT_DIR, f"shirt_{i}.png")
    generate_shirt_image(image_path, pattern, color, material, ai_texture)

    if i % 100 == 0:
        print(f"Generated {i} images...")

print("✅ Dataset Generation Complete!")
```

```python

import os
import cv2
import requests
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options

# =============================
# 🔹 Step 1: Set Up Image Scraper
# =============================

def scrape_google_images(query, num_images=20):
    """Scrape Google Images for pattern designs."""
    search_url = f"https://www.google.com/search?tbm=isch&q={query.replace(' ', '+')}"
    headers = {"User-Agent": "Mozilla/5.0"}

    response = requests.get(search_url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    image_urls = []
    for img in soup.find_all("img")[1:num_images+1]:  # Skip first placeholder image
        if "src" in img.attrs:
            image_urls.append(img["src"])

    return image_urls

# =============================
# 🎨 Step 2: Extract Dominant Colors
# =============================

def extract_colors(image_path, num_colors=5):
    """Extract dominant colors from an image."""
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = image.reshape((-1, 3))

    unique_colors, counts = np.unique(image, axis=0, return_counts=True)
    sorted_indices = np.argsort(-counts)[:num_colors]
    dominant_colors = unique_colors[sorted_indices]

    return dominant_colors

# =============================
# 📥 Step 3: Download and Process Images
# =============================

query = "abstract pattern design"
os.makedirs("pattern_dataset", exist_ok=True)

# Scrape image URLs
image_urls = scrape_google_images(query, num_images=20)

for i, img_url in enumerate(image_urls):
    response = requests.get(img_url, stream=True)
    image_path = f"pattern_dataset/pattern_{i}.jpg"

    with open(image_path, "wb") as f:
        f.write(response.content)

    # Extract colors
    colors = extract_colors(image_path)
    print(f"🎨 Pattern {i}: Dominant Colors → {colors}")

print(f"✅ Collected {len(image_urls)} pattern designs with color data!")
```

```python

import os
import cv2
import time
import requests
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# =============================
# 🔹 Step 1: Set Up Web Scraper
# =============================

def scrape_google_images(query, num_images=500):
    """Scrape Google Images dynamically with scrolling."""
    options = Options()
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

    search_url = f"https://www.google.com/search?tbm=isch&q={query.replace(' ', '+')}"
    driver.get(search_url)

    image_urls = set()
    last_height = driver.execute_script("return document.body.scrollHeight")

    while len(image_urls) < num_images:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

        imgs = driver.find_elements(By.TAG_NAME, "img")
        for img in imgs:
            src = img.get_attribute("src")
            if src and "http" in src:
                image_urls.add(src)
            if len(image_urls) >= num_images:
                break

        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break  # Stop if no new images are loaded
        last_height = new_height

    driver.quit()
    return list(image_urls)

# =============================
# 🎨 Step 2: Extract Dominant Colors
# =============================

def extract_colors(image_path, num_colors=5):
    """Extract dominant colors from an image."""
    image = Image.open(image_path).convert("RGB")
    image = image.resize((100, 100))
    image_array = np.array(image).reshape((-1, 3))

    unique_colors, counts = np.unique(image_array, axis=0, return_counts=True)
    sorted_indices = np.argsort(-counts)[:num_colors]
    dominant_colors = unique_colors[sorted_indices]

    return dominant_colors

# =============================
# 📥 Step 3: Download and Process Images
# =============================

query = "psychedelic pattern design"
os.makedirs("pattern_dataset", exist_ok=True)

# Scrape large dataset of images
image_urls = scrape_google_images(query, num_images=500)

for i, img_url in enumerate(image_urls):
    try:
        response = requests.get(img_url, stream=True, timeout=10)
        image_path = f"pattern_dataset/pattern_{i}.jpg"

        with open(image_path, "wb") as f:
            f.write(response.content)

        # Extract colors
        colors = extract_colors(image_path)
        print(f"🎨 Pattern {i}: Dominant Colors → {colors}")

    except Exception as e:
        print(f"⚠️ Skipping image {i}: {e}")

print(f"✅ Collected {len(image_urls)} pattern designs with color data!")
```
