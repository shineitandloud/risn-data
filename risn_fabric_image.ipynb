```python
!pip install diffusers transformers accelerate safetensors --quiet
from diffusers import StableDiffusionPipeline
import torch
import os

# Check if GPU is available (Mobile users should use CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load AI model (uses less memory)
pipeline = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1",
    torch_dtype=torch.float16
).to(device)

# Set output folder
output_dir = "/content/generated_fabrics"
os.makedirs(output_dir, exist_ok=True)

# Generate Images
for i in range(10):  # Adjust batch count based on mobile limitations
    prompt = "High-tech futuristic fabric pattern, ultra HD, 8K"
    image = pipeline(prompt).images[0]
    image.save(f"{output_dir}/fabric_{i}.png")
    print(f"Saved: fabric_{i}.png")

print(f"All images saved in {output_dir}")
```
